```python
import ipywidgets as widgets
import csv
```

```python
w1 = widgets.Combobox()
w2 = widgets.Combobox()
isl1 = widgets.IntSlider(min=0, max=10, step=1, value=5)
hbox1 = widgets.HBox([w1,w2, isl1])
widgets.VBox([hbox1,hbox1])
```

```python
class Material:
    def __init__(self, name, density):
        self.name = name
        self.density = density
```

```python
with open('construction_material_densities_eu.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    data = [Material(row['material_lv'], float(row['density_kg_m3_typical'])) for row in reader]
```

```python
print([material.name for material in data])
```

```python
dropdown = widgets.Dropdown(
    options=[material.name for material in data])
```
