```python
import ipywidgets as widgets
from IPython.display import display, clear_output

profile_name_output = widgets.Output()

def profile_name_changed(change):
    with profile_name_output:
        clear_output(wait=True)
        print("Profile changed to:", change["new"])

sections.observe(profile_name_changed, names="selected_profile")
display(profile_name_output)
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, Math

L = widgets.FloatText(value=6.0, description="L [m]")
q = widgets.FloatText(value=10.0, description="q [kN/m]")
E = widgets.FloatText(value=210.0, description="E [GPa]")
deflection_output = widgets.Output()

def update_deflection(change=None):
    profile = sections.selected_profile_data
    I_y = profile["I_y_mm4"]
    L_mm = L.value * 1000
    q_n_per_mm = q.value
    E_n_per_mm2 = E.value * 1000
    delta_mm = 5 * q_n_per_mm * L_mm**4 / (384 * E_n_per_mm2 * I_y)

    with deflection_output:
        clear_output(wait=True)
        display(Math(r"\delta_{max}=\frac{5qL^4}{384EI}"))
        print(f"Profile: {profile['profile']}")
        print(f"I_y = {I_y:,.0f} mm^4")
        print(f"Deflection = {delta_mm:.2f} mm")

sections.observe(update_deflection, names="selected_profile_data")
L.observe(update_deflection, names="value")
q.observe(update_deflection, names="value")
E.observe(update_deflection, names="value")

display(widgets.HBox([L, q, E]), deflection_output)
update_deflection()
```
